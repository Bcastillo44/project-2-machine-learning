```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import lag_plot
from pandas import datetime
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
att_combined_path = Path('Resources/combined_price_inc_data/att_master.csv')
att_combined = pd.read_csv(att_combined_path, parse_dates=True, infer_datetime_format=True)

#df = pd.read_csv("T.csv")
att_combined.head(5)
```

```python
#df.shape
att_combined.shape
```

```python
plt.figure()
#lag_plot(df['Open'], lag=3)
lag_plot(att_combined['Close'], lag=3)
plt.title('AT&T Stock - Autocorrelation plot with lag = 3')
plt.show()
```

```python
#plt.plot(df.index, df["Close"])
plt.plot(att_combined["Date"], att_combined["Close"])
plt.xticks(np.arange(0,9332, 200), att_combined["Date"][0:9332:200])
plt.title("AT&T stock price over time")
plt.xlabel("time")
plt.ylabel("price")
plt.show()
```

```python
#train_data, test_data = df[0:int(len(df)*0.7)], df[int(len(df)*0.7):]
train_data, test_data = att_combined[0:int(len(att_combined)*0.7)], att_combined[int(len(att_combined)*0.7):]
training_data = train_data['Close'].values
test_data = test_data['Close'].values
history = [x for x in training_data]
model_predictions = []
N_test_observations = len(test_data)
for time_point in range(N_test_observations):
    model = ARIMA(history, order=(4,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    model_predictions.append(yhat)
    true_test_value = test_data[time_point]
    history.append(true_test_value)
MSE_error = mean_squared_error(test_data, model_predictions)
print('Testing Mean Squared Error is {}'.format(MSE_error))
```

```python
train_data.shape
```

```python
test_set_range = att_combined[int(len(att_combined)*0.7):].index
plt.figure(figsize=(20,10))
plt.plot(test_set_range[-188:], model_predictions[-188:], color='blue',label='Predicted Price')
plt.plot(test_set_range[-188:], test_data[-188:], color='red', label='Actual Price')
plt.title('AT&T Prices Prediction (MSE:{})'.format(MSE_error))
plt.xlabel('Date')
plt.ylabel('Prices')
plt.xticks(np.arange(9144,9332,50), att_combined.Date[9144:9332:50])
plt.legend()
plt.show()
```
